```python
import seaborn as sn
plt.figure(figsize = (30, 5))
Specialist_count = df['Specialist'].value_counts()
ax = sn.barplot(x=Specialist_count.index,y=Specialist_count.values, palette='PuBu_r')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.savefig('Specialist Ranking.png', bbox_inches='tight')
plt.show()
```
