Importing libraries for testiong and training.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

import math
```

Assigning obeservation data and predictable data the variables

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,131]
```

Examining the dimensions of the data

```python
print(X.shape)
print(Y.shape)
print(X.info())
print(Y.info())
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

Splitting the training and testing data.

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

<h3>Random Forest
<H5>Training

```python
clfRF=RandomForestClassifier(criterion='entropy',min_samples_leaf=2)
clfRF.fit(X_train,Y_train)
```

<h3>Testing 

```python
Y_predRF=clfRF.predict(X_test)
```

```python
Y_predRF[0:20]
```

```python
Y_test[0:20]
```

As seen here they are almost 100% same

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
print(confusion_matrix(Y_test, Y_predRF))
print(classification_report(Y_test, Y_predRF))
print("Train Accuracy(RF): ", accuracy_score(Y_train, clfRF.predict(X_train)))
print("Test Accuracy(RF): ", accuracy_score(Y_test, Y_predRF))
```

<h6>The confusion matrix you've provided is a square matrix where each row and column represents a specific class or medical condition. It shows the counts of true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN) for each class in your classification problem.

Here's a breakdown of what this confusion matrix might represent:

Rows represent the actual classes or medical conditions.
Columns represent the predicted classes or medical conditions.
The numbers in the matrix indicate the counts of instances falling into different categories.
For example, let's break down the first row:

Row 1: It appears to correspond to the first class or medical condition (e.g., 'Malaria').
The first column (column 1) shows 28 instances where 'Malaria' was correctly predicted ('Malaria' true positives).
All other columns in row 1 (columns 2 onwards) contain zeros, which means there were no false positives for 'Malaria,' no cases where other conditions were incorrectly predicted as 'Malaria.'
You can interpret the other rows in a similar manner. Each row represents a different medical condition, and the columns indicate the predictions made by your classification model.

If you have specific questions about the performance of your model or need further analysis of the confusion matrix, please let me know.

With a binary confusion matrix, you can calculate various metrics to evaluate the performance of your disease prediction model, including:

Accuracy: (TP + TN) / (TP + TN + FP + FN)<br>
Precision: TP / (TP + FP)<br>
Recall (Sensitivity or True Positive Rate): TP / (TP + FN)<br>
Specificity (True Negative Rate): TN / (TN + FP)<br>
F1 Score: 2 * (Precision * Recall) / (Precision + Recall)</h6>

<H3>Multi-Layer Perception
<H5>Training

```python
epochs=10
nos_batch_size=32
max_iters=epochs*(len(X_train)//nos_batch_size)
clfMLP=MLPClassifier(activation='relu',batch_size=nos_batch_size,hidden_layer_sizes=(64,32),max_iter=max_iters)
clfMLP.fit(X_train,Y_train)
```

Thus we have trained our MLP model with 10 epochs and a batch size of 32 while max iterations throught out are ~ 1240 with 2 hidden layers each of 64 and 32 nodes respectively.

<H5>Testing

```python
Y_predMLP=clfMLP.predict(X_test)
```

```python
Y_predMLP[0:20]
```

```python
# data to plot
n_groups = 2
algorithms = ('Random Forest (RF)','Multilayer Perceptron (MLP) Neural Network' )
train_accuracy = (accuracy_score(Y_train, clfRF.predict(X_train))*100,
                  accuracy_score(Y_train, clfMLP.predict(X_train))*100)
test_accuracy = (accuracy_score(Y_test, Y_predRF)*100, 
                accuracy_score(Y_test, Y_predMLP)*100)

# create plot
fig, ax = plt.subplots(figsize=(15, 5))
index = np.arange(n_groups)
bar_width = 0.3
opacity = 0.8
rects1 = plt.bar(index, train_accuracy, bar_width, alpha = opacity, color='Cornflowerblue', label='Train')
rects2 = plt.bar(index + bar_width, test_accuracy, bar_width, alpha = opacity, color='Teal', label='Test')
plt.xlabel('Algorithm') # x axis label
plt.ylabel('Accuracy (%)') # y axis label
plt.ylim(0, 115)
plt.title('Comparison of Algorithm Accuracies') # plot title
plt.xticks(index + bar_width * 0.5, algorithms) # x axis data labels
plt.legend(loc = 'upper right') # show legend
for index, data in enumerate(train_accuracy):
    plt.text(x = index - 0.035, y = data + 1, s = round(data, 2), fontdict = dict(fontsize = 8))
for index, data in enumerate(test_accuracy):
    plt.text(x = index + 0.25, y = data + 1, s = round(data, 2), fontdict = dict(fontsize = 8))
plt.savefig('Algorithm Accuracies.png')
plt.show()
```
